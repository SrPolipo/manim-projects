```python
class Vicsek_Example(MovingCameraScene):
    def construct(self):
        vicsek_instance = vicsek(num_points = 20000, colors = [BLUE,WHITE,BLUE,RED,RED], stroke_width = 3, noise = 0.02)
        self.add(vicsek_instance)
        self.camera.frame.set(height = vicsek_instance.height*1.4)
        self.wait()
        for _ in range(7):
            self.play(vicsek_instance.animate.iterate())
            self.wait(0.3)
        self.wait(2)
```

```python
class Barnsley_Example(MovingCameraScene):
    def construct(self):
        barnsley_instance = barnsley(num_points = 40000, colors = [GREEN], stroke_width = 2, noise = 0.06)
        self.add(barnsley_instance)
        self.camera.frame.shift(3*UP).scale(1.4)
        self.wait()
        for i in range(20):
            self.play(barnsley_instance.animate.iterate(), self.camera.frame.animate.shift(0.4/(i+1)*UP), run_time = 0.7, rate_func = rate_functions.linear)
        self.wait(2)
```

```python
class Sierpinski_Example(MovingCameraScene):
    def construct(self):
        sierpinski_instance = sierpinski_triangle(num_points = 40000, colors = [RED, GREEN, BLUE], stroke_width = 6, noise = 0.03)
        self.add(sierpinski_instance)
        self.camera.frame.set(height=sierpinski_instance.height*1.4)
        self.wait()
        for _ in range(7):
            self.play(sierpinski_instance.animate.iterate())
            self.wait(0.3)
        self.wait(2)
```
